```python
from youtube_transcript_api import YouTubeTranscriptApi, TranscriptsDisabled
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import PromptTemplate
from langchain_community.document_loaders import WebBaseLoader,PyMuPDFLoader,UnstructuredPDFLoader
```

# **Step 1 : Indexing**

Step 1A : Indexing - Document Ingestion

```python
# here i have taken 4 types of input one is copy-paste of all information from website store into a string variable and work,second is using the youtube videos of tradebulls third is direct using tradebulls website url and fourth can directly upload pdfs like the daily research pdf

text = """
FAQs on Opening Demat and Trading Accounts for Persons with Disabilities
The following FAQs are in line with SEBI's circulars, guidelines, and FAQs

Can individuals with disabilities open demat and trading accounts with Tradebulls Securities Private Limited?

→ Yes, Tradebulls Securities Private Limited allows individuals with disabilities to open both demat and trading accounts.


Is a guardian required for individuals with disabilities to open a demat or trading account with Tradebulls Securities Private Limited?

→ Adults (18 years and above) who are of sound mind and not legally disqualified from entering into contracts can open accounts independently. If a guardian is needed as per SEBI's FAQs, Tradebulls Securities Private Limited accepts a Guardianship Certificate issued by the Local Level Committee, in line with the Mental Health Care Act, 2017, and the National Trust for the Welfare of Persons with Autism, Cerebral Palsy, Mental Retardation and Multiple Disabilities Act, 1999. Both the guardian and the individual with disabilities must comply with all applicable KYC norms.


How are persons with disabilities categorised as minors or majors?

→ Classification is based on age. Those 18 years and older are considered majors if they are of sound mind and not legally disqualified to contract. SEBI has separate guidelines on the opening of demat and trading accounts for minors available on their website.


What form is needed for persons with disabilities to open a demat or trading account with Tradebulls Securities Private Limited?

→ As per SEBI's guidelines, uniform documentation is required for all trading account openings, including for persons with disabilities. This includes a Know Your Client (KYC) form which captures basic details of the client and an instruction/checklist for filling it out. You must also submit certification under the Rights of Persons with Disabilities Act, 2016.


What if an individual with disabilities cannot sign the forms themselves? Do both the guardian and the individual need to comply with KYC norms?

→ Tradebulls Securities Private Limited provides multiple different avenues for individuals to sign the forms themselves including through physical signatures. In the event that an individual is still unable to open the account, the account can be opened with the guardian’s signature. Both the guardian and the individual with disabilities must comply with all applicable KYC norms.


Does Tradebulls Securities Private Limited offer online/digital KYC for persons with disabilities?

→ Yes, Tradebulls Securities Private Limited provides online/digital KYC. Our team will assist clients in facilitating their KYC, if requested.


How does a person with disabilities initiate the process of opening a demat and trading account with Tradebulls Securities Private Limited?

→ To initiate the process of opening a demat and trading account with Tradebulls Securities Private Limited, persons with disabilities would need to contact our customer support team. The first step is to write to our customer support team via email at wecare@tradebulls.com or call our customer care number at 079-40001000, informing us of your intent to open an account. Our sensitised dedicated support team will then provide personalised assistance and guide you through each step of the process.

We are committed to ensuring a smooth and accessible onboarding experience, and our team is here to help address any questions or concerns you may have throughout the process.


How does Tradebulls Securities Private Limited address difficulties faced by persons with disabilities during the onboarding process, specifically when required to?

→ Tradebulls Securities Private Limited has implemented several solutions to assist clients with disabilities during the onboarding process, particularly for the signature requirement:
Clients have the option to provide a physical signature on all documents which shall be signed in presence of the authorized representative of Tradebulls Securities Private Limited, which can then be mailed to Tradebulls Securities Private Limited.

"""

url = "https://www.tradebulls.in/blogs/personal-finance/how-to-protect-yourself-from-stock-market-scams"

loader = WebBaseLoader(url)
docs = loader.load()

loader2 = UnstructuredPDFLoader("daily_research_pdf.pdf",mode="elements")
docs2 = loader2.load()

pdf_docs = []

for doc in docs2:
  pdf_docs.append(doc.page_content)

pdf_docs
```

```python
video_id = "e9B2F3m0CL8"

try:
    # Initialize API (new version)
    api = YouTubeTranscriptApi()

    # Fetch transcript (auto-detects best language)
    transcript_list = api.fetch(video_id)

    # Convert to clean text
    transcript = " ".join(
        chunk.text.replace("\n", " ").strip()
        for chunk in transcript_list
        if len(chunk.text.strip()) > 2
    )

    print("Transcript extracted successfully!\n")
    print(len(transcript))
    print(transcript)

except TranscriptsDisabled:
    print("No captions available for this video.")

except Exception as e:
    print("Error:", str(e))
```

Step 1B : Indexing - Text Splitting

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=700,
    chunk_overlap=100,
    separators=["\n\n", "\n", "→", ".", " "]
)

chunks = splitter.create_documents([text,transcript,docs[0].page_content] + pdf_docs)
```

```python
len(chunks)
```

```python
chunks[4]
```

Step 1C & D : Indexing - Embedding Generation and Storing in Vector Store

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
vector_store = FAISS.from_documents(chunks, embeddings)
```

```python
vector_store.index_to_docstore_id
```

# **Step 2 : Retrival**

```python
retriever = vector_store.as_retriever(search_type="similarity",search_kwargs={"k":4})
```

```python
retriever
```

```python
retriever.invoke("Is a guardian required for individuals with disabilities to open a demat or trading account with Tradebulls Securities Private Limited?")
```

# **Step 3 : Augmentation**

```python
llm = ChatOpenAI(model="gpt-4o-mini",temperature=0.2)
```

```python
prompt = PromptTemplate(
    template="""
      You are a helpful assistant for an online trading platform.
      Answer ONLY from the provided text context.
      If the context is insufficient, just say you don't know but in a proper text message.

      {context}
      Question: {question}
    """,
    input_variables = ['context', 'question']
)
```

```python
question          = "What is the date of the report"
retrieved_docs    = retriever.invoke(question)
```

```python
context_text = "\n\n".join(doc.page_content for doc in retrieved_docs)
```

```python
final_prompt = prompt.invoke({"context": context_text, "question": question})
```

```python
final_prompt
```

# **Step 4 : Generation**

```python
response = llm.invoke(final_prompt)
print(response.content)
```

# **Building a Chain for RAG**

```python
from langchain_core.runnables import RunnableParallel, RunnablePassthrough, RunnableLambda
from langchain_core.output_parsers import StrOutputParser
```

```python
def format_docs(retrieved_docs):
  context_text = "\n\n".join(doc.page_content for doc in retrieved_docs)
  return context_text
```

```python
parallel_chain = RunnableParallel(
    context=RunnableLambda(format_docs),
    question=RunnablePassthrough()
)
```

```python
rag_chain = (
    RunnableParallel({
        "context": retriever | RunnableLambda(format_docs),
        "question": RunnablePassthrough()
    })
    | prompt
    | llm
    | StrOutputParser()
)
final_chain = rag_chain
```

```python
response = final_chain.invoke("Explain the breakdown pattern in the 14th May report in easy way?")
print(response)
```
